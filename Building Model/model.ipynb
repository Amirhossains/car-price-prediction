```python
import pandas as pd
import numpy as np
import pickle as pk
import json as js
import warnings
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split,\
ShuffleSplit, cross_val_score, GridSearchCV, KFold
from sklearn.linear_model import LinearRegression, Lasso
```

```python
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('../database/cleaned_database.csv')
```

```python
df.head(20)
```

```python
df.shape
```

```python
df.rename(columns={'transmission': 'is_manual'}, inplace=True)
```

```python
for i in range(df.shape[0]):
    if df['is_manual'][i] == 'Manual':
        df['is_manual'][i] = 1
    else:
        df['is_manual'][i] = 0
```

```python
df.head(20)
```

```python
for i in range(df.shape[0]):
    if df['owner'][i] == 'First Owner':
        df['owner'][i] = 1
    elif df['owner'][i] == 'Second Owner':
        df['owner'][i] = 2
    elif df['owner'][i] == 'Third Owner':
        df['owner'][i] = 3
    else:
         df['owner'][i] = 4
```

```python
df.head(20)
```

```python
dummies = pd.get_dummies(df.name)
```

```python
dummies.head()
```

```python
df1 = pd.concat([df.drop('name', axis='columns'), dummies], axis='columns') 
```

```python
df1.head()
```

```python
X = df1.drop('selling_price', axis='columns')
y = df1.selling_price
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    X, y, test_size=0.05, random_state=20
)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
lr.score(x_test, y_test)
```

```python
cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
cross_val_score(estimator=LinearRegression(), X=X, y=y, cv=cv)
```

```python
models = {
    'LinearRegression': {
        'model': LinearRegression(),
        'params': {
            'positive': [False, True],
            'n_jobs': [1, 3, 6, 9, 12]
        }
    },
    'DecisionTreeClassifier': {
        'model': DecisionTreeClassifier(),
        'params': {
            'splitter': ['best', 'random'],
            'criterion': ['gini', 'log_loss', 'entropy']
        }
    },
    'Lasso': {
        'model': Lasso(),
        'params': {
            'alpha': [1.0, 2.0],
            'selection': ['random', 'cyclic']
        }
    }
}
```

```python
cv = KFold(n_splits=5, shuffle=True, random_state=0)
```

```python
scores = []
for model_name, mp in models.items():
    G = GridSearchCV(estimator=mp['model'], param_grid=mp['params'],
                     cv=cv, return_train_score=False, error_score='raise')
    G.fit(X, y)
    scores.append({
        'model_name': model_name,
        'best_score': G.best_score_,
        'best_params': G.best_params_
    })
```

```python
GRResult = pd.DataFrame(scores, columns=['model_name','best_score','best_params'])
```

```python
GRResult
```

```python
model = LinearRegression(n_jobs=1, positive=False)
```

```python
model.fit(X, y)
model.score(X, y)
```

```python
def predict_price_of_car(name, year, km_driven, is_manual, owner, seats, max_power, Mileage):
    name_column_index = np.where(X.columns==name)[0][0]
    x = np.zeros(len(X.columns))
    x[0] = year
    x[1] = km_driven
    x[2] = is_manual
    x[3] = owner
    x[4] = seats
    x[5] = max_power
    x[6] = Mileage
    x[name_column_index] = 1
    return model.predict([x])[0]
```

```python
print(predict_price_of_car('Maruti', 2014, 145500, 1, 1, 5, 74, 23.40))
print(predict_price_of_car('Hyundai', 2017, 45000, 1, 1, 5, 81.86, 20.14))
print(predict_price_of_car('Toyota', 2011, 90000, 1, 1, 5, 67.10, 23.59))
print(predict_price_of_car('Ford', 2013, 169000, 1, 1, 5, 68.10, 20.00))
```
